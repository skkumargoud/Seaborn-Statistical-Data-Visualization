```python
import seaborn as sns 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#laod the data avaible in the default dataset avaible in seaborn
tips=sns.load_dataset("tips")
tips.head()
```

# #Regression Plots 

# lm.plot() is used to plot the regression line  in seaborn 

```python
sns.lmplot(y="tip",x='total_bill',data=tips) 
```

```python
#by using hue arugument  categorical variable 
sns.lmplot(y="tip",x='total_bill',data=tips,hue="sex")
```

```python
#changes the scatter size
sns.lmplot(y="tip",x='total_bill',data=tips,hue='sex',markers=["*","+"],scatter_kws={"s":70})
```

```python
#by using col and row
sns.lmplot(y="tip",x='total_bill',data=tips,hue='sex',col="sex",row="time")
```

```python
#changing the aspect ratio
sns.lmplot(y="tip",x='total_bill',data=tips,hue='sex',col="day",aspect=0.6,size=5)
```
