```python
import seaborn as sns 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#laod the data avaible in the default dataset avaible in seaborn
tips=sns.load_dataset("tips")
tips.head()
```

# barplot

```python
sns.barplot(x=("sex"),y=("total_bill"),data=tips)
```

```python
#by default they will do the average if u are changing the other aggreate function by estimater argument
sns.barplot(x=("sex"),y=("total_bill"),data=tips,estimator=np.sum)
```

## Countplot()it mainly count of occurances of categorical data values


```python
sns.countplot(x=("sex"),data=tips)
```

```python
sns.countplot(x=("sex"),hue=("smoker"),data=tips)
```

# Boxplot() it tell the categorical data distribution  in quadrille  show the outlier in the data 


```python
sns.boxplot(x=("day"),y=("total_bill"),data=tips)
```

```python
#by using hue arugument add another categorical variable 
sns.boxplot(x=("day"),y=("total_bill"),data=tips,hue=('smoker'))
```

# violinplot() it is very similar to box-plot() it is used to distribute the data categorical variable


```python
sns.violinplot(x=("day"),y=("total_bill"),data=tips)
```

```python
#by using hue arugument add another categorical variable 
sns.violinplot(x=("day"),y=("total_bill"),data=tips,hue=('smoker'))
```

```python
#by using hue arugument add another categorical variable 
sns.violinplot(x=("day"),y=("total_bill"),data=tips,hue=('smoker'))
```

```python
#by using thesplit  arugument  by using the plot the data side by side
sns.violinplot(x=("day"),y=("total_bill"),data=tips,hue=('smoker'),split=True)
```

# stripplot() it is used  to compare the relationship between one categorical variable and numberic variable it is just like scatter plot


```python
sns.stripplot(x=("day"),y=("total_bill"),data=tips)
```

```python
#by using hue arugument add another categorical variable 
sns.stripplot(x=("day"),y=("total_bill"),data=tips,hue=('smoker'))
```

```python
#by using thesplit  arugument  by using the plot the data side by side
sns.stripplot(x=("day"),y=("total_bill"),data=tips,hue=('smoker'),split=True)
```

## SwarnPlot() It’s a combination of both violin plot and strip plot 


```python
sns.swarmplot(x="day",y='total_bill',data=tips)
```

```python
sns.violinplot(x=("day"),y=("total_bill"),data=tips)
sns.swarmplot(x="day",y='total_bill',data=tips,color="black")
```
